```python
import pandas as pd
logins = pd.read_csv('logins.csv')
apps   = pd.read_csv('applications.csv')
users  = pd.read_csv('users.csv')
logins = logins.merge(users, on='user_id')
apps   = apps.merge(users, on='user_id')
```

```python
apps['date'] = pd.to_datetime(apps['date'])
logins['date'] = pd.to_datetime(logins['date'])
```

```python
def prevpost(date):
    if date > pd.to_datetime('2024-07-01'):
        return "Post"
    else:
        return "Pre"
    
apps['PvP'] = apps['date'].apply(prevpost)
logins['PvP'] = logins['date'].apply(prevpost)
```

```python
logins['year'] = logins['date'].dt.year
logins = logins[logins['year']==2024]
```

```python
apps['weekly'] = apps['date'].dt.isocalendar()['week']
apps['monthly'] = apps['date'].dt.month

logins['weekly'] = logins['date'].dt.isocalendar()['week']
logins['monthly'] = logins['date'].dt.month
```

```python
from scipy import stats

apps_pre = apps[apps['PvP']=='Pre'].groupby('date')['application_id'].count()
apps_post = apps[apps['PvP']=='Post'].groupby('date')['application_id'].count()
t_stat, p_val = stats.ttest_ind(apps_post, apps_pre, equal_var=False)
t_stat, p_val
```

```python

apps_pre = apps[apps['PvP']=='Pre'].groupby('user_id')['application_id'].count()
apps_post = apps[apps['PvP']=='Post'].groupby('user_id')['application_id'].count()
t_stat, p_val = stats.ttest_ind(apps_post, apps_pre, equal_var=False)
t_stat, p_val
```

```python

apps_pre = apps[apps['PvP']=='Pre'].groupby('school').agg(u =('user_id','nunique'),a =('application_id','count')).reset_index()
apps_post = apps[apps['PvP']=='Post'].groupby('school').agg(u =('user_id','nunique'),a =('application_id','count')).reset_index()
```

```python
apps_pre['apps_per_user_pre'] = apps_pre['a'] / apps_pre['u']
apps_post['apps_per_user_post'] = apps_post['a'] / apps_post['u']
```

```python
df= pd.merge(apps_pre,apps_post,on='school')
df['lift'] = round(((df['apps_per_user_post'] - df['apps_per_user_pre']) / df['apps_per_user_pre'])*100,2).astype(str) + '%'
df
```

```python
for school in df['school'].unique():
    pre = apps[(apps.PvP=='Pre') & (apps.school==school)].groupby('user_id').size()
    post= apps[(apps.PvP=='Post')& (apps.school==school)].groupby('user_id').size()
    t,p = stats.ttest_ind(post, pre, equal_var=False)
    print(school, 'lift=', post.mean()/pre.mean()-1, 't=',t, 'p=',p)
```
